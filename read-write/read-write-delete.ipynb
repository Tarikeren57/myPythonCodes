```python
# Dosya yolunu belirtin
dosya_yolu = 'test.txt'  # Buraya dosyanızın tam yolunu yazın

# Yazmak istediğiniz metni belirleyin
eklenecek_yazi = "I am from Türkiye."

# Dosyayı açıp yazma işlemi
with open(dosya_yolu, 'a', encoding='utf-8') as file:
    file.write("\n" + eklenecek_yazi  )  # \n yeni satıra geçmek için kullanılır

print("Yazı dosyaya eklendi.")
```

```python
def dosya_oku(dosya_yolu):
    """Dosyanın içeriğini okur ve ekrana yazdırır."""
    try:
        with open(dosya_yolu, 'r', encoding='utf-8') as file:
            print("Dosya İçeriği:")
            print(file.read())
    except FileNotFoundError:
        print("Dosya bulunamadı!")

def dosya_sil(dosya_yolu):
    """Dosyanın içeriğini siler (boşaltır)."""
    try:
        with open(dosya_yolu, 'w', encoding='utf-8') as file:
            file.write("")  # Dosya içeriğini boşalt
        print("Dosyanın içeriği silindi.")
    except FileNotFoundError:
        print("Dosya bulunamadı!")

def dosyaya_yazi_ekle(dosya_yolu, yazi):
    """Dosyaya yeni yazı ekler."""
    try:
        with open(dosya_yolu, 'a', encoding='utf-8') as file:
            file.write(yazi + "\n")
        print("Yeni yazı dosyaya eklendi.")
    except FileNotFoundError:
        print("Dosya bulunamadı!")

def menu():
    """Kullanıcıya seçim yapması için bir menü sunar ve işlemi gerçekleştirir."""
    dosya_yolu = 'test.txt'  # Dosya yolunu belirtin
    
    while True:
        print("\n--- Menü ---")
        print("1. Dosya içeriğini oku")
        print("2. Dosya içeriğini sil")
        print("3. Dosyaya yazı ekle")
        print("4. Çıkış")
        
        secim = input("Bir seçenek girin (1/2/3/4): ")
        
        if secim == '1':
            dosya_oku(dosya_yolu)
        elif secim == '2':
            dosya_sil(dosya_yolu)
        elif secim == '3':
            yazi = input("Eklenecek yazıyı girin: ")
            dosyaya_yazi_ekle(dosya_yolu, yazi)
        elif secim == '4':
            print("Çıkılıyor...")
            break
        else:
            print("Geçersiz seçenek, lütfen tekrar deneyin.")

# Menü fonksiyonunu çalıştırmak için:
menu()
```

```python
def dosya_oku(dosya_yolu):
    """Dosyanın içeriğini okur ve ekrana yazdırır."""
    try:
        with open(dosya_yolu, 'r', encoding='utf-8') as file:
            print("\nDosya İçeriği:")
            print(file.read())
    except FileNotFoundError:
        print("Dosya bulunamadı! Lütfen doğru dosya yolunu girin.")
    except Exception as e:
        print(f"Bir hata oluştu: {e}")

def dosya_sil(dosya_yolu):
    """Dosyanın içeriğini siler (boşaltır)."""
    try:
        with open(dosya_yolu, 'w', encoding='utf-8') as file:
            file.write("")  # Dosya içeriğini boşalt
        print("\nDosyanın içeriği silindi.")
    except FileNotFoundError:
        print("Dosya bulunamadı! Lütfen doğru dosya yolunu girin.")
    except Exception as e:
        print(f"Bir hata oluştu: {e}")

def dosyaya_yazi_ekle(dosya_yolu, yazi):
    """Dosyaya yeni yazı ekler."""
    try:
        with open(dosya_yolu, 'a', encoding='utf-8') as file:
            file.write(yazi + "\n")
        print("\nYeni yazı dosyaya eklendi.")
    except FileNotFoundError:
        print("Dosya bulunamadı! Lütfen doğru dosya yolunu girin.")
    except Exception as e:
        print(f"Bir hata oluştu: {e}")

def menu():
    """Kullanıcıya seçim yapması için bir menü sunar ve işlemi gerçekleştirir."""
    
    # Dosya yolunu kullanıcıdan al
    dosya_yolu = input("Lütfen dosya yolunu girin: ")
    
    while True:
        print("\n--- Menü ---")
        print("1. Dosya içeriğini oku")
        print("2. Dosya içeriğini sil")
        print("3. Dosyaya yazı ekle")
        print("4. Çıkış")
        
        secim = input("Bir seçenek girin (1/2/3/4): ")
        
        if secim == '1':
            dosya_oku(dosya_yolu)
        elif secim == '2':
            dosya_sil(dosya_yolu)
        elif secim == '3':
            yazi = input("Eklenecek yazıyı girin: ")
            dosyaya_yazi_ekle(dosya_yolu, yazi)
        elif secim == '4':
            print("Çıkılıyor...")
            break
        else:
            print("Geçersiz seçenek, lütfen tekrar deneyin.")

# Menü fonksiyonunu çalıştırmak için:
menu()
```
